```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u


# --- パラメータ ---
El_fixed = 30.0 * u.deg  # deg
Az_values = np.linspace(0, 360, 361) * u.deg # 0～360 deg

delta_Az = []

for Az in Az_values:
    # 順変換
    Az_offset, El_offset = pe.apply_offset(Az, El_fixed)
    
    # 逆変換
    Az_rec, El_rec = pe.apply_inverse_offset(Az_offset, El_offset)
    
    # 差分
    delta_Az.append((Az_rec - Az).to_value(u.arcsec))

delta_Az = np.array(delta_Az)

# --- プロット ---
plt.figure(figsize=(10,5))
plt.plot(Az_values, delta_Az)  # arcsec単位に変換したい場合
plt.xlabel("Input Az [deg]")
plt.ylabel("Az_rec - Az [arcsec]")
plt.title(f"Offset + Inverse Offset check (El={El_fixed})")
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u

# --- パラメータ ---
El_list = [20.0, 40.0, 60.0]  # deg
Az_values = np.linspace(0, 360, 361) * u.deg

plt.figure(figsize=(10,6))

for El_fixed in El_list:
    delta_Az = []
    for Az in Az_values:
        # 順変換
        Az_offset, El_offset = pe.apply_offset(Az, El_fixed*u.deg)
        # 逆変換
        Az_rec, El_rec = pe.apply_inverse_offset(Az_offset, El_offset)
        # 差分（arcsec単位）
        delta_Az.append((Az_rec - Az).to(u.arcsec).value)

    
    delta_Az = np.array(delta_Az)
    plt.plot(Az_values, delta_Az, label=f'El={El_fixed} deg')

plt.xlabel("Input Az [deg]")
plt.ylabel("Az_rec - Az [arcsec]")
plt.title("Offset + Inverse Offset check for different El")
plt.grid(True)
plt.legend()
plt.show()
```
